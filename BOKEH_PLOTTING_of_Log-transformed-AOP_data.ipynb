```python
dfs = df[['mxbl-gr', 'mxbl2gr', 'r411_2_r555', 'r411_toRed', 
          'mxblID', 'mxBlColor', 'provCode', 'provColor', 'biome', 'biomColor',
          'blueBiome','blBiomColor','blueProvince', 'blProvColor', 'chl_l']]
```

```python
def modify_category_plot(doc):
    def make_new_source(cat_list):
        filtered_df = dfs[dfs[selection_label].isin(cat_list)]
        return ColumnDataSource(filtered_df)

    def make_checkboxgroup():
        categories = dfs[selection_label].unique().tolist()
        cbg_selection = CheckboxGroup(labels=categories,
                                     active=[0, 1])
        return cbg_selection

    def update(attr, old, new):
        # get the list of selections to graph
        selections_to_plot = [selection.labels[i] for i in selection.active]

        # make a new dataset based on the selection and the make_dataset function
        new_source = make_new_source(selections_to_plot)
        # Update the source used in the plot
        source.data.update(new_source.data)
        
    def make_plot(src):
        biome_selection = CheckboxGroup(labels=df.biome.unique().tolist(), )
        p = figure(plot_width=600, plot_height=600,
                   title=plot_title, x_axis_label=x_lbl,
                   y_axis_label='log(chl)')
        p.circle(x=x_data_col_name, y='chl_l', fill_color=color_dict[selection_label],
                 size=10, line_color='black',source=src,)
        hover = HoverTool(tooltips=cur_tool_tip)
        p.add_tools(hover)
        return p
    
    selection = make_checkboxgroup()
    selection.on_change('active', update)
    initial_selection = [selection.labels[i] for i in selection.active]
    source = make_new_source(initial_selection)
    p = make_plot(source)
    control=WidgetBox(selection, width=120)
    layout = row(control, p)
    doc.add_root(layout)
```

```python
plot_title = 'OC4'
x_lbl = 'log(max(Blue)) - log(Green)'
x_data_col_name = 'mxbl-gr'
cur_tool_tip = [#('biome', '@biome'),
                ('maxBlue', '@mxblID'),('province', '@provCode')]
selection_label = 'biome'
color_dict = dict(biome='biomColor', )
source = ColumnDataSource(dfs)
handler = FunctionHandler(modify_category_plot)
app = Application(handler)
show(app)
```
